```python
import yfinance as yf
import statsmodels.api as sm
```

```python
aapl = yf.Ticker('AAPL')
data = aapl.history(start='2025-01-01')
data.head()
```

```python
data['Return'] = data['Close'].pct_change()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.head(5)
```

```python
# tommorows close
data['Target Close'] = data['Close'].shift(-1)
```

```python
data
```

```python
data["Target Returns"] = data["Return"].shift(-1)
```

```python
data
```

```python
data.dropna(axis=0, inplace=True)
```

# Closing Price Regression

```python
model = sm.OLS(data["Target Close"], sm.add_constant(data['Close']))
r = model.fit()
print(r.summary())
```

# Return Regression

```python
model = sm.OLS(data["Target Returns"], sm.add_constant(data['Return']))
r = model.fit()
print(r.summary())
```

# Why is this the case?
- Price process of a stock has a **unit root**.
- So its not stationary, we can not apply typical models.
- Use transformations to get rid of the unit root of the time series.
